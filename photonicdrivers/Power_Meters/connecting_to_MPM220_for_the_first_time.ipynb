# How to find the MAC address of a mystery device

1️⃣ Physically connect it

    Use an Ethernet cable from the device directly to your computer.

2️⃣ Give your PC a static IP

Since you don’t know the device’s IP, you need your PC in the same general range to even “see” it.

On Windows:

    - Go to Control Panel → Network and Internet → Network and Sharing Center → Change adapter settings. Right-click Ethernet → Properties → Internet Protocol Version 4 (TCP/IPv4).

    Select Use the following IP address:
        IP: 192.168.1.10 (arbitrary, but not .1)
        Subnet mask: 255.255.255.0
        Gateway: leave blank.
        Click OK.

3️⃣ Scan common static IP ranges

    Many devices ship with factory default IPs like:

        192.168.0.x
        192.168.1.x
        192.168.100.x
        10.0.0.x

Run nmap for each:

    sudo nmap -sn 192.168.1.0/24
    sudo nmap -sn 192.168.0.0/24
    sudo nmap -sn 192.168.100.0/24
    sudo nmap -sn 10.0.0.0/24
